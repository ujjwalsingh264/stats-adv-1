#Question 1 : What are Type I and Type II errors in hypothesis testing, and how do they impact decision-making?

#ANSWER 1 --
Type I and Type II errors are fundamental concepts in hypothesis testing, crucial for understanding the reliability of statistical conclusions. Here's a breakdown:

#Type I Error (False Positive):
This occurs when you reject a true null hypothesis. In simpler terms, you conclude there is a significant effect or difference when, in reality, there isn't. The probability of making a Type I error is denoted by α (alpha), also known as the significance level. A common alpha level is 0.05 (or 5%), meaning there's a 5% chance of incorrectly rejecting a true null hypothesis.

Impact on decision-making: A Type I error can lead to acting on a non-existent effect. For example:

   In medicine, releasing a drug that is mistakenly thought to be effective, potentially causing harm or wasting resources.

   In manufacturing, halting a production line to fix a problem that doesn't actually exist, leading to unnecessary downtime and costs.

#Type II Error (False Negative):
 This occurs when you fail to reject a false null hypothesis. In simpler terms, you conclude there is no significant effect or difference when, in reality, there is one. The probability of making a Type II error is denoted by β (beta). The power of a test (1 - β) is the probability of correctly rejecting a false null hypothesis.

   Impact on decision-making: A Type II error can lead to missing a real effect or difference. For example:

     In medicine, failing to identify an effective drug, potentially delaying life-saving treatment.

     In manufacturing, continuing a faulty process because the test failed to detect the problem, leading to defective products.

#Trade-off between Type I and Type II Errors:

There's an inverse relationship between Type I and Type II errors. Decreasing the probability of one typically increases the probability of the other (for a fixed sample size). The choice of α (significance level) reflects which type of error is considered more costly or undesirable in a given context. For instance:

  High cost for Type I error (e.g., medical diagnoses): A lower α is preferred to minimize false positives, even if it increases the risk of Type II errors.

  High cost for Type II error (e.g., security screening for dangerous items): A higher α might be accepted to minimize false negatives, ensuring that truly dangerous items are detected, even if it leads to more false alarms.

Understanding these errors helps researchers and decision-makers make informed choices about how to design studies, interpret results, and manage the risks associated with statistical inference.



Question 2: What is the P-value in hypothesis testing, and how should it be interpreted in the context of the null hypothesis?


The P-value is a crucial concept in hypothesis testing that helps us decide whether to reject or fail to reject a null hypothesis.

#What is the P-value?
The P-value, or probability value, is the probability of observing a test statistic as extreme as, or more extreme than, the one calculated from your sample data, assuming that the null hypothesis is true. In simpler terms, it quantifies the strength of the evidence against the null hypothesis.

#How to Interpret the P-value in the Context of the Null Hypothesis:

#a.Compare P-value to Significance Level (α):

Before conducting a hypothesis test, you define a significance level (α), which is the maximum probability of making a Type I error (rejecting a true null hypothesis) that you are willing to accept. Common alpha values are 0.05 (5%), 0.01 (1%), or 0.10 (10%).

#b.Decision Rule:

If P-value ≤ α: This indicates that the observed data is unlikely to have occurred by random chance if the null hypothesis were true. Therefore, you reject the null hypothesis. This suggests that there is statistically significant evidence to support the alternative hypothesis.

If P-value > α: This indicates that the observed data is reasonably likely to have occurred by random chance if the null hypothesis were true. Therefore, you fail to reject the null hypothesis. This means there is not enough statistically significant evidence to conclude that the alternative hypothesis is true.

#Important Considerations:

 * P-value is NOT the probability that the null hypothesis is true. It's the probability of the data given the null hypothesis.

  * A low P-value doesn't necessarily mean a large or important effect. It only indicates statistical significance. The practical significance of the findings should also be considered.

 Failing to reject the null hypothesis does NOT mean the null hypothesis is true. It simply means there isn't enough evidence from the sample data to reject it at the chosen significance level.

 **Context is Key:** Always interpret the P-value within the context of your research question, experimental design, and the potential consequences of Type I and Type II errors.


#Question 3:Explain the difference between a Z-test and a T-test, including when to use each.

# ANSWER 3 :

 Certainly! Both Z-tests and T-tests are statistical hypothesis tests used to compare means. The primary difference between them lies in the information you have about the population standard deviation and the sample size. Here's a breakdown:

#Z-test
 A  Z-test is a statistical test used to determine whether two population means are different when the variances are known, and the sample size is large, or when comparing a sample mean to a population mean with a known population standard deviation.

#Key Characteristics:

Population Standard Deviation (σ): Must be known.
Sample Size (n): Typically used when the sample size is large (generally n > 30), even if the population standard deviation is unknown (due to the Central Limit Theorem, the sample standard deviation can approximate the population standard deviation with large samples).
Distribution: Assumes the sampling distribution of the mean is normally distributed.
#When to Use a Z-test:

Comparing a sample mean to a population mean when the population standard deviation is known. (e.g., You know the average height of all adult males in a country and their standard deviation, and you want to see if a specific sample of adult males is significantly different).
Comparing two sample means when both population standard deviations are known.
With large sample sizes (n > 30) when the population standard deviation is unknown (in this case, the sample standard deviation can be used as an estimate for the population standard deviation).
Example Scenario: A company claims its light bulbs have an average lifespan of 1,000 hours with a known population standard deviation of 50 hours. You take a sample of 100 light bulbs and find their average lifespan to be 990 hours. You would use a Z-test to determine if the sample's lifespan is significantly different from the company's claim.

#T-test
What it is: A T-test is a statistical test used to determine if there is a significant difference between the means of two groups, or if a sample mean is significantly different from a hypothesized population mean, especially when the population standard deviation is unknown and the sample size is small.

#Key Characteristics:

Population Standard Deviation (σ): Unknown.
Sample Size (n): Typically used with small sample sizes (generally n < 30). For larger samples, the T-distribution approaches the Z-distribution.
Distribution: Uses the t-distribution, which accounts for the extra uncertainty introduced by estimating the population standard deviation from the sample.
Degrees of Freedom: The t-distribution's shape depends on the degrees of freedom (usually n-1 for a single sample t-test), which impacts the critical values.
#When to Use a T-test:

Comparing a sample mean to a population mean when the population standard deviation is unknown.
Comparing the means of two independent samples when the population standard deviations are unknown. (e.g., Comparing the test scores of two different teaching methods).
Comparing the means of two dependent (paired) samples when the population standard deviations are unknown. (e.g., Comparing a patient's blood pressure before and after treatment).
Example Scenario: A researcher wants to test if a new fertilizer increases plant growth. They apply the fertilizer to a small sample of 15 plants and compare their average growth to the average growth of plants without fertilizer. The population standard deviation of plant growth is unknown. A T-test would be appropriate here.

#Summary of Differences:
Feature	Z-test	T-test
Population Std. Dev. (σ)	Known	Unknown
Sample Size (n)	Large (n > 30) or any size if σ is known	Small (n < 30) or any size if σ is unknown
Distribution Used	Standard Normal Distribution (Z-dist)	Student's t-Distribution (t-dist)
Uncertainty Factor	Less uncertainty (σ known)	More uncertainty (σ estimated from sample)
In essence, if you have sufficient information about the population (specifically, its standard deviation) and/or a large sample size, a Z-test is often suitable. If you have to estimate the population standard deviation from your sample, especially with a small sample, a T-test is the more appropriate and robust choice because it accounts for the additional variability due to that estimation.





#Question 4 : What is a confidence interval, and how does the margin of error influence its width and interpretation?

#ANSWER 4 -
 A confidence interval is a range of values, derived from sample data, that is likely to contain the true value of an unknown population parameter. It provides an estimated range of plausible values for an unknown parameter (e.g., a population mean or proportion) rather than a single point estimate.

* Key Components of a Confidence Interval:
Point Estimate: This is the single best guess for the population parameter, usually calculated from the sample data (e.g., sample mean, sample proportion).
Margin of Error: This is the range around the point estimate that creates the interval. It accounts for the sampling variability and the desired level of confidence.
* Confidence Level expresses the probability that the interval will contain the true population parameter. Common confidence levels are 90%, 95%, or 99%. A 95% confidence interval means that if you were to take many samples and construct a confidence interval from each, about 95% of those intervals would contain the true population parameter.
The general formula for a confidence interval is:

Confidence Interval = Point Estimate ± Margin of Error

 # The Margin of Error

The margin of error quantifies the precision of the estimate. It is the maximum expected difference between the sample statistic (point estimate) and the true population parameter. It is typically calculated as:

Margin of Error = Critical Value × Standard Error of the Statistic

* Critical Value This value depends on the chosen confidence level and the sampling distribution (e.g., Z-score for large samples or known population standard deviation, t-score for small samples or unknown population standard deviation).

* Standard Error: This measures the variability of the sample statistic (e.g., standard deviation of the sampling distribution of the mean).

* the Margin of Error Influences Its Width and Interpretation:
The margin of error directly influences the width of the confidence interval:

* Larger Margin of Error → Wider Confidence Interval: A larger margin of error means there is more uncertainty in your estimate, leading to a broader range of plausible values for the population parameter. This makes the statement about the true parameter less precise.

* Smaller Margin of Error → Narrower Confidence Interval: A smaller margin of error indicates greater precision in your estimate, resulting in a tighter range of values. This makes the statement about the true parameter more precise.
Factors Affecting the Margin of Error (and thus the width of the CI):
Confidence Level:

* Higher Confidence Level (e.g., 99% vs. 95%): Requires a larger critical value, which increases the margin of error and thus widens the confidence interval. To be more confident that you've captured the true parameter, you need a broader net.

* Lower Confidence Level (e.g., 90% vs. 95%): Requires a smaller critical value, which decreases the margin of error and narrows the confidence interval. You are less confident, so you can afford a tighter range.
Sample Size (n):

* Larger Sample Size: Reduces the standard error (as it's often inversely proportional to the square root of n), which decreases the margin of error and narrows the confidence interval. More data generally leads to more precise estimates.

* Smaller Sample Size: Increases the standard error, which increases the margin of error and widens the confidence interval. Less data means more uncertainty.
Population Standard Deviation (σ) or Variability:

Higher Population Standard Deviation: Indicates more variability in the population, which increases the standard error, leading to a larger margin of error and a wider confidence interval. If the data points are very spread out, your estimate will naturally be less precise.

Lower Population Standard Deviation: Indicates less variability, which decreases the standard error, leading to a smaller margin of error and a narrower confidence interval. If the data points are clustered closely, your estimate can be more precise.

Interpretation of a Confidence Interval:
If you have a 95% confidence interval for the average height of adult males as (68 inches, 72 inches), it means:

We are 95% confident that the true average height of adult males in the population falls between 68 and 72 inches.
It does NOT mean there is a 95% probability that the true mean is in this specific interval. Instead, it reflects the reliability of the estimation procedure over many repeated samples.
In summary, the margin of error is a crucial component that dictates the precision and breadth of your confidence interval. Understanding its relationship with confidence level, sample size, and population variability is key to correctly interpreting statistical estimates.



#Question 5: Describe the purpose and assumptions of an ANOVA test. How does it extend hypothesis testing to more than two groups?


 # ANSWER 5

 Here's an explanation of the purpose, assumptions, and extension of ANOVA for hypothesis testing with more than two groups:

#What is an ANOVA test?

ANOVA stands for Analysis of Variance. It's a collection of statistical models and their associated estimation procedures (such as the variation among and between groups) used to analyze the differences among group means in a sample. ANOVA was developed by Ronald Fisher.

#Purpose of an ANOVA Test:

The primary purpose of an ANOVA test is to determine if there are any statistically significant differences between the means of three or more independent (unrelated) groups. While you could perform multiple t-tests to compare pairs of groups, ANOVA offers a more efficient and statistically robust solution.

Specifically, ANOVA tests the null hypothesis (H₀) that the means of all groups are equal against the alternative hypothesis (H₁), which states that at least one group mean is different from the others.

* H₀ (Null Hypothesis): μ₁ = μ₂ = μ₃ = ... = μₖ (All group means are equal)

* H₁ (Alternative Hypothesis): At least one group mean is different from the others.

#Assumptions of an ANOVA Test:

For the results of an ANOVA test to be reliable, several assumptions should be met:

a. Independence of Observations: The observations within each group, and between groups, must be independent of each other. This means that the measurement of one subject does not influence the measurement of another subject.

b. Normality: The dependent variable (the variable being measured) should be approximately normally distributed for each group. While ANOVA is somewhat robust to minor deviations from normality, especially with larger sample sizes, severe non-normality can affect the results.

c. Homogeneity of Variances (Homoscedasticity): The variance of the dependent variable should be approximately equal across all groups. This means that the spread of data around the mean should be similar for all groups being compared. Tests like Levene's test or Bartlett's test can be used to check this assumption.

d. Categorical Independent Variable(s) and Continuous Dependent Variable: The independent variable (the grouping variable) must be categorical (e.g., different treatment groups, different types of fertilizer), and the dependent variable must be continuous (interval or ratio scale, e.g., height, test scores, yield).

#How ANOVA Extends Hypothesis Testing to More Than Two Groups:

If you have more than two groups and you want to compare their means, running multiple independent t-tests can lead to a problem known as Type I error inflation (also known as the family-wise error rate).

Here's why:

* When you perform a single hypothesis test with a significance level (α) of, say, 0.05, there's a 5% chance of incorrectly rejecting a true null hypothesis (a Type I error).

* If you perform multiple t-tests, say to compare three groups (Group A vs. B, A vs. C, B vs. C), you are doing three separate tests. The probability of making at least one Type I error across these multiple tests increases significantly (it's no longer just 0.05). For 'k' comparisons, the family-wise error rate is approximately 1 - (1 - α)ᵏ.

#ANOVA solves this problem by performing a single, omnibus test.

Instead of comparing each pair of means individually, ANOVA assesses the overall variation within the data. It partitions the total variability into two main components:

a. Between-Group Variability (MSB - Mean Square Between): This measures the variation between the means of the different groups. It reflects the effect of the independent variable.

b. Within-Group Variability (MSW - Mean Square Within): This measures the variation among the individual observations within each group. It represents random error or unexplained variance.

ANOVA calculates an F-statistic, which is the ratio of the between-group variability to the within-group variability:

#F = MSB / MSW

* If the F-statistic is large: This suggests that the variability between the group means is significantly larger than the variability within the groups, indicating that at least one group mean is different.

* If the F-statistic is small (close to 1): This suggests that the between-group variability is similar to the within-group variability, meaning there's likely no significant difference between the group means.

#Post-Hoc Tests:

If the ANOVA test yields a statistically significant F-statistic, it tells you that a difference exists somewhere among the group means, but it doesn't tell you which specific groups are different from each other. To find these specific differences, you would then conduct post-hoc tests (e.g., Tukey's HSD, Bonferroni correction, Scheffé's test). These tests are designed to control the family-wise error rate when making multiple comparisons.

In summary, ANOVA provides an efficient way to test for mean differences across multiple groups while controlling the Type I error rate, making it a powerful tool for comparing more than two conditions or populations.



# Question 6: Write a Python program to perform a one-sample Z-test and interpret the result for a given dataset.

# ANSWER 6:
### One-Sample Z-test in Python

This program will demonstrate how to perform a one-sample Z-test using Python's `statsmodels` library. A one-sample Z-test is used to compare a sample mean to a known population mean when the population standard deviation is known.

**Scenario:** Let's say we know the average weight of adult males in a city is 75 kg with a standard deviation of 10 kg. We take a sample of 30 adult males from a particular neighborhood and want to see if their average weight is significantly different from the city's average.

```python
import numpy as nP

from scipy import stats # Import scipy for normal distribution functions
# from statsmodels.stats.weightstats import ztest # ztest from statsmodels does not support known population_std directly

# 1. Generate a sample dataset (or use your own data)
# Let's assume our sample of 30 males has a mean slightly different from the population mean
np.random.seed(42) # for reproducibility
sample_data = np.random.normal(loc=78, scale=10, size=30)

# 2. Define population parameters (known values)
population_mean = 75 # kg (hypothesized population mean under the null hypothesis)
population_std = 10 # kg (known population standard deviation)
sample_size = len(sample_data)
sample_mean = np.mean(sample_data)

# 3. Perform the one-sample Z-test (manual calculation as ztest from statsmodels doesn't take known population_std)
# Calculate the Z-statistic
# Z = (sample_mean - population_mean) / (population_std / sqrt(sample_size))
z_statistic = (sample_mean - population_mean) / (population_std / np.sqrt(sample_size))

# Calculate the p-value for a two-tailed test
# For a two-tailed test, we use the survival function (1 - CDF) and multiply by 2
p_value = 2 * stats.norm.sf(abs(z_statistic))

# 4. Interpret the results
significance_level = 0.05 # Alpha value

print(f"Sample Mean: {sample_mean:.2f} kg")
print(f"Hypothesized Population Mean: {population_mean} kg")
print(f"Known Population Standard Deviation: {population_std} kg")
print(f"Number of Samples: {sample_size}")
print("----------------------------------------")
print(f"Calculated Z-statistic: {z_statistic:.3f}")
print(f"P-value: {p_value:.3f}")
print(f"Significance Level (alpha): {significance_level}")
print("----------------------------------------")

if p_value < significance_level:
    print(f"Since the P-value ({p_value:.3f}) is less than the significance level ({significance_level}),")
    print("we reject the null hypothesis.")
    print(f"Conclusion: There is statistically significant evidence that the sample mean ({sample_mean:.2f} kg) is different from the hypothesized population mean ({population_mean} kg).")
else:
    print(f"Since the P-value ({p_value:.3f}) is greater than the significance level ({significance_level}),")
    print("we fail to reject the null hypothesis.")
    print(f"Conclusion: There is not enough statistically significant evidence to conclude that the sample mean ({sample_mean:.2f} kg) is different from the hypothesized population mean ({population_mean} kg).")
```

#Question 8: Generate multiple samples from a non-normal distribution and implement the Central Limit Theorem using Python.


#ANSWER 8:

### Central Limit Theorem (CLT) Demonstration

#### What is the Central Limit Theorem?

The Central Limit Theorem (CLT) is a fundamental theorem in statistics that states that, given a sufficiently large sample size, the sampling distribution of the sample mean will be approximately normally distributed, regardless of the shape of the population distribution. This holds true as long as the samples are independent and identically distributed, and the population has a finite variance.

Key takeaways:
1.  **Shape:** Even if the original population distribution is non-normal (e.g., exponential, uniform, skewed), the distribution of sample means will tend towards a normal (bell-shaped) distribution.
2.  **Mean:** The mean of the sampling distribution of the means will be approximately equal to the population mean.
3.  **Standard Deviation:** The standard deviation of the sampling distribution of the means (known as the standard error) will be equal to the population standard deviation divided by the square root of the sample size.

#### How we'll demonstrate it:

We will:
1.  Choose a non-normal distribution, for example, the **Exponential Distribution**.
2.  Generate a large number of samples from this exponential distribution.
3.  For each sample, calculate its mean.
4.  Plot a histogram of these sample means. According to the CLT, this histogram should resemble a normal distribution.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# 1. Define Population Parameters (Exponential Distribution - Non-Normal)
# The exponential distribution is often used to model the time until an event occurs.
# Its probability density function (PDF) is skewed.
population_scale = 2 # This is 1/lambda. So, lambda (rate parameter) = 1/2 = 0.5
                     # For exponential, population mean (mu) = scale, population std = scale
population_mean = population_scale
population_std = population_scale

# 2. Define Simulation Parameters
num_samples = 10000    # Number of samples to draw from the population
sample_size = 30       # Size of each individual sample (must be 'sufficiently large' for CLT)

# 3. Generate Multiple Samples and Calculate their Means
sample_means = []
for _ in range(num_samples):
    # Draw a sample of 'sample_size' from the exponential distribution
    sample = np.random.exponential(scale=population_scale, size=sample_size)
    # Calculate the mean of this sample
    sample_means.append(np.mean(sample))

# Convert to numpy array for easier calculations
sample_means = np.array(sample_means)

# 4. Plot the Distribution of Sample Means
plt.figure(figsize=(12, 7))

# Histogram of sample means
plt.hist(sample_means, bins=30, density=True, alpha=0.7, color='skyblue', edgecolor='black',
         label=f'Distribution of Sample Means (N={sample_size})')

# Overlay a theoretical normal distribution (as predicted by CLT)
# CLT predicts: mean of sample means = population mean
#             : std dev of sample means (standard error) = population_std / sqrt(sample_size)
theoretical_mean = population_mean
theoretical_std_error = population_std / np.sqrt(sample_size)

xmin, xmax = plt.xlim()
x = np.linspace(xmin, xmax, 100)
p = norm.pdf(x, theoretical_mean, theoretical_std_error)
plt.plot(x, p, 'r', linewidth=2, label=f'Theoretical Normal Dist.\n(Mean={theoretical_mean:.2f}, Std Error={theoretical_std_error:.2f})')

# Add titles and labels
plt.title('Demonstration of the Central Limit Theorem with Exponential Distribution')
plt.xlabel('Sample Mean')
plt.ylabel('Density')
plt.legend()
plt.grid(True, linestyle=':', alpha=0.6)
plt.axvline(x=population_mean, color='green', linestyle='--', label=f'Population Mean ({population_mean:.2f})')
plt.legend()
plt.show()

print(f"\nObserved Mean of Sample Means: {np.mean(sample_means):.3f}")
print(f"Theoretical Mean (Population Mean): {population_mean:.3f}")
print(f"Observed Standard Deviation of Sample Means: {np.std(sample_means):.3f}")
print(f"Theoretical Standard Error (Population Std / sqrt(sample_size)): {theoretical_std_error:.3f}")
```

#Question 9: Write a Python function to calculate and visualize the confidence interval for a sample mean.

#Answer 9:

### Confidence Interval Calculation and Visualization

This program defines a Python function to calculate and visualize the confidence interval for a sample mean. The function accounts for whether the population standard deviation is known or unknown, using the appropriate statistical distribution (Z-distribution or t-distribution).

#### What is a Confidence Interval?
A confidence interval provides a range of plausible values for an unknown population parameter (like the population mean) based on sample data. For example, a 95% confidence interval means that if you were to repeat the sampling process many times, 95% of the intervals you construct would contain the true population mean.

#### How it works:
1.  **Input:** Sample data, desired confidence level (e.g., 0.95 for 95%), and optionally, the known population standard deviation.
2.  **Calculation:**
    *   It determines whether to use a Z-interval (if `population_std` is provided) or a T-interval (if `population_std` is `None`).
    *   It calculates the sample mean, standard error, and the critical values corresponding to the chosen confidence level.
    *   It then computes the lower and upper bounds of the confidence interval.
3.  **Visualization:**
    *   A histogram of the `sample_data` is plotted to show the distribution of the individual data points.
    *   The sample mean is marked with a vertical line.
    *   The calculated confidence interval is displayed as a horizontal line or shaded region, centered around the sample mean.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats

def calculate_and_visualize_confidence_interval(
    sample_data: np.ndarray,
    confidence_level: float = 0.95,
    population_std: float = None
):
    """
    Calculates and visualizes the confidence interval for a sample mean.

    Args:
        sample_data (np.ndarray): The 1D array of sample data.
        confidence_level (float): The desired confidence level (e.g., 0.95 for 95%).
        population_std (float, optional): The known population standard deviation.
                                           If None, the sample standard deviation is used
                                           and a t-interval is calculated. Defaults to None.

    Returns:
        tuple: A tuple containing the (lower_bound, upper_bound) of the confidence interval.
    """
    if not (0 < confidence_level < 1):
        raise ValueError("Confidence level must be between 0 and 1.")

    n = len(sample_data)
    if n < 2:
        raise ValueError("Sample size must be at least 2 to calculate a confidence interval.")

    sample_mean = np.mean(sample_data)

    # Determine whether to use Z-distribution or t-distribution
    if population_std is not None:
        # Z-interval: population standard deviation is known
        statistic_type = "Z-interval (Population Std Known)"
        standard_error = population_std / np.sqrt(n)
        lower_bound, upper_bound = stats.norm.interval(confidence_level, loc=sample_mean, scale=standard_error)
    else:
        # T-interval: population standard deviation is unknown
        statistic_type = "T-interval (Population Std Unknown)"
        sample_std = np.std(sample_data, ddof=1) # ddof=1 for sample standard deviation
        standard_error = sample_std / np.sqrt(n)
        df = n - 1 # Degrees of freedom
        lower_bound, upper_bound = stats.t.interval(confidence_level, df=df, loc=sample_mean, scale=standard_error)

    # --- Visualization ---
    plt.figure(figsize=(10, 6))
    plt.hist(sample_data, bins=15, density=True, alpha=0.6, color='skyblue', edgecolor='black',
             label='Sample Data Distribution')

    plt.axvline(sample_mean, color='red', linestyle='--', label=f'Sample Mean: {sample_mean:.2f}')

    # Plot the confidence interval
    plt.hlines(0, lower_bound, upper_bound, colors='green', linestyles='-', lw=4,
               label=f'{int(confidence_level*100)}% CI: [{lower_bound:.2f}, {upper_bound:.2f}]')
    plt.plot([lower_bound, lower_bound], plt.ylim(), color='green', linestyle=':', lw=1)
    plt.plot([upper_bound, upper_bound], plt.ylim(), color='green', linestyle=':', lw=1)

    plt.title(f'Confidence Interval for Sample Mean ({statistic_type})')
    plt.xlabel('Value')
    plt.ylabel('Density / Frequency')
    plt.legend()
    plt.grid(axis='y', alpha=0.75)
    plt.show()

    print(f"\n{int(confidence_level*100)}% Confidence Interval: ({lower_bound:.2f}, {upper_bound:.2f})")
    print(f"Sample Mean: {sample_mean:.2f}")
    print(f"Standard Error: {standard_error:.2f}")

    return lower_bound, upper_bound

# --- Example Usage ---
print("\n--- Example 1: T-interval (Population Std Unknown) ---")
np.random.seed(42)
sample_data_t = np.random.normal(loc=50, scale=8, size=30) # Small sample, std unknown
calculate_and_visualize_confidence_interval(sample_data_t, confidence_level=0.95)

print("\n--- Example 2: Z-interval (Population Std Known) ---")
np.random.seed(42)
population_true_std = 7 # Assume we know the population standard deviation
sample_data_z = np.random.normal(loc=52, scale=population_true_std, size=100) # Larger sample
calculate_and_visualize_confidence_interval(sample_data_z, confidence_level=0.90, population_std=population_true_std)

print("\n--- Example 3: T-interval with different confidence level ---")
np.random.seed(123)
sample_data_t2 = np.random.exponential(scale=10, size=25) # Non-normal distribution, small sample
calculate_and_visualize_confidence_interval(sample_data_t2, confidence_level=0.99)
```

### Chi-square Goodness-of-Fit Test

#### What is the Chi-square Goodness-of-Fit Test?

The Chi-square goodness-of-fit test is a statistical hypothesis test used to determine if a sample data's observed frequencies (counts) match the expected frequencies derived from a specific theoretical distribution or a prior hypothesis. In simpler terms, it assesses how well an observed distribution fits an expected distribution.

**Hypotheses:**
*   **Null Hypothesis (H₀):** The observed distribution does not differ significantly from the expected distribution (i.e., the data fits the hypothesized distribution).
*   **Alternative Hypothesis (H₁):** The observed distribution differs significantly from the expected distribution (i.e., the data does not fit the hypothesized distribution).

#### How it works:

1.  **Observed Frequencies:** These are the actual counts from your sample data for different categories.
2.  **Expected Frequencies:** These are the counts you would expect in each category if the null hypothesis were true (i.e., if your data perfectly matched the theoretical distribution or prior expectation).
3.  **Test Statistic (χ²):** The test calculates a chi-square statistic based on the differences between observed and expected frequencies. Larger differences lead to a larger χ² value.
4.  **P-value:** The p-value is then calculated from the χ² statistic and the degrees of freedom. A small p-value (typically ≤ 0.05) indicates that the observed deviations from the expected distribution are unlikely to have occurred by chance, leading to the rejection of the null hypothesis.

#### Example Scenario:

Let's imagine a classic example: a die roll. If a die is fair, we expect each face (1 through 6) to appear an equal number of times over many rolls. We'll simulate rolling a die 120 times and then see if the observed frequencies of each face match the expected frequencies for a fair die.

```python
import numpy as np
from scipy import stats

# 1. Define Observed Frequencies
# Let's simulate 120 die rolls and their outcomes.
# For a fair die, we'd expect 20 rolls for each face (120 / 6 = 20).
# We'll create some slightly skewed observed frequencies.

np.random.seed(42) # For reproducibility
# Simulate 120 rolls of a die, slightly biased towards higher numbers
observed_frequencies = np.array([15, 18, 22, 25, 20, 20])
# Let's adjust to ensure the sum is exactly 120 if needed, or create more randomly
# For demonstration, let's ensure it sums to 120 manually if not generated directly
# observed_frequencies = np.random.randint(low=10, high=30, size=6) # More random example
# observed_frequencies = observed_frequencies / observed_frequencies.sum() * 120
# observed_frequencies = np.round(observed_frequencies).astype(int)

# Ensure the sum matches the total number of trials for expected_frequencies calculation
total_trials = observed_frequencies.sum()

# 2. Define Expected Frequencies (for a fair 6-sided die)
# If the die is fair, each face should appear with equal probability.
# Total trials / number of faces
num_categories = len(observed_frequencies)
expected_frequencies = np.array([total_trials / num_categories] * num_categories)

print(f"Observed Frequencies: {observed_frequencies}")
print(f"Expected Frequencies (for a fair die): {expected_frequencies}")
print("----------------------------------------")

# 3. Perform the Chi-square Goodness-of-Fit Test
# stats.chisquare returns the chi-square statistic and the p-value
chi2_statistic, p_value = stats.chisquare(f_obs=observed_frequencies, f_exp=expected_frequencies)

# 4. Interpret the results
significance_level = 0.05 # Alpha value

print(f"Chi-square Statistic: {chi2_statistic:.3f}")
print(f"P-value: {p_value:.3f}")
print(f"Significance Level (alpha): {significance_level}")
print("----------------------------------------")

if p_value < significance_level:
    print(f"Since the P-value ({p_value:.3f}) is less than the significance level ({significance_level}),")
    print("we reject the null hypothesis.")
    print("Conclusion: There is statistically significant evidence that the observed distribution is significantly different from the expected distribution (i.e., the die is likely not fair).")
else:
    print(f"Since the P-value ({p_value:.3f}) is greater than the significance level ({significance_level}),")
    print("we fail to reject the null hypothesis.")
    print("Conclusion: There is not enough statistically significant evidence to conclude that the observed distribution is different from the expected distribution (i.e., the die appears to be fair).")
```
